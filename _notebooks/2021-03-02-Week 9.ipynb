We'll make some test data ready to use for the learner

```python
path = untar_data(URLs.MNIST_SAMPLE)

threes = (path/'train'/'3').ls().sorted()
sevens = (path/'train'/'7').ls().sorted()

seven_tensors = [tensor(Image.open(o)) for o in sevens]
three_tensors = [tensor(Image.open(o)) for o in threes]

stacked_sevens = torch.stack(seven_tensors).float()/255
stacked_threes = torch.stack(three_tensors).float()/255
```

## Assignment 1: I skal bygge jeres egen learner fra bunden af ved at lave en klasse I Python der hedder MyLearner eller noget i den stil. Learneren skal indeholde de 7 punkter bogen beskriver på side 150, og som gennemgås af to omgange fra hhv. side 159-162 og igen side 171-176.

```python
class MyLearner:
    def __init__(self, dls, model, opt_func, loss_func, metrics):
        self.dls = dls
        self.model = model
        self.opt_func = opt_func
        self.loss_func = loss_func
        self.metrics = metrics
        
        
        self.train_loss = 0
        self.valid_loss = 0
        self.batch_accuracy = 0
    
    def fit(self, n_pochs, lr):
        for i in range(n_pochs):
            self.train_epoch()
            print(f"epoch: {i}; train_loss: {self.train_loss}; valid_loss: {self.valid_loss}; batch_accuracy: {self.batch_accuracy};")
            
    def train_epoch(self):
        for xb,yb in self.dls.train:
            self.calc_grad(xb, yb)
```

```python
train_x = torch.cat([stacked_threes, stacked_sevens]).view(-1, 28*28)
train_y = tensor([1]*len(threes) + [0]*len(sevens)).unsqueeze(1)
dset = list(zip(train_x,train_y))
dl = DataLoader(dset, batch_size=256, num_workers=0)

valid_3_tens = torch.stack([tensor(Image.open(o)) 
                            for o in (path/'valid'/'3').ls()])
valid_3_tens = valid_3_tens.float()/255
valid_7_tens = torch.stack([tensor(Image.open(o)) 
                            for o in (path/'valid'/'7').ls()])
valid_7_tens = valid_7_tens.float()/255

valid_x = torch.cat([valid_3_tens, valid_7_tens]).view(-1, 28*28)
valid_y = tensor([1]*len(valid_3_tens) + [0]*len(valid_7_tens)).unsqueeze(1)
valid_dset = list(zip(valid_x,valid_y))
valid_dl = DataLoader(valid_dset, batch_size=256)

dls = DataLoaders(dl, valid_dl)

batch = train_x[:4]

def init_params(size, std=1.0):
    return (torch.randn(size)*std).requires_grad_()

weights = init_params((28*28,1))
bias = init_params(1)

def linear1(xb):
    return xb@weights + bias

preds = linear1(batch)

def mnist_loss(predictions, targets):
    return torch.where(targets==1, 1-predictions, predictions).mean()

def batch_accuracy(xb, yb):
    preds = xb.sigmoid()
    correct = (preds>0.5) == yb
    return correct.float().mean()
```

```python
learn = MyLearner(dls, nn.Linear(28*28,1), opt_func=SGD, loss_func=mnist_loss, metrics=batch_accuracy)

learn.fit(10, lr=1.)
```
